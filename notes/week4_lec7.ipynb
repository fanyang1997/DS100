```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from numpy import sin, cos, exp, sqrt, pi, linspace, arange
```

```python
x = linspace(0, 2 * pi)
y = sin(x)
plt.plot(x, y, label='sin(x)')
plt.legend()
plt.title('Harmonic')
plt.xlabel('x')
plt.ylabel('y')

z = cos(x)
plt.plot(x, z, label='cos(x)')

plt.figure()
plt.plot(x, sin(2*x), label='sin(2x)')
plt.legend()
```

```python
f, ax = plt.subplots()
ax.plot(x, y, label='sin(x)')
ax.legend()
ax.set_title('Harmonic')
ax.set_xlabel('x')
ax.set_ylabel('y')

f1, ax1 = plt.subplots()
ax1.plot(x, sin(2*x), label='sin(2x)')
ax1.legend()

ax.plot(x, z, label='cos(x)')
```

```python
x = linspace(0, 2*pi, 400)
y = sin(x**2)

f, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.scatter(x, y)

f.suptitle('Two plots')
```

```python
f = plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))

for ax in f.axes:
    for t in ax.get_xticklabels() + ax.get_yticklabels():
        t.set_visible(False)
        
f.suptitle('Subplot2grid')

ax3.plot([1, 2, 3])
```
